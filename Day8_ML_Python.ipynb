```python
# Simple Linear Regression
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Input data
x = np.array([5,15,25,35,45,55]).reshape((-1,1))
y = np.array([5,20,14,32,22,38])
```

```python

# Train model
model = LinearRegression().fit(x, y)
print("R^2 Score:", model.score(x, y))
print("Intercept:", model.intercept_)
print("Slope:", model.coef_)
```

```python
# Predictions
y_pred = model.predict(x)
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='green')
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Predict new values
x_new = np.arange(5).reshape((-1,1))
y_new = model.predict(x_new)
print("Predicted new Y values:", y_new)
```

```python
# Salary Prediction Example
import pandas as pd
salary_data = pd.read_csv("Salary_Data.csv")
x = salary_data.iloc[:, :-1].values
y = salary_data.iloc[:, -1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)

regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Predicting
x_pred = regressor.predict(x_train)
y_pred = regressor.predict(x_test)
```

```python
# Visualization - Train
plt.scatter(x_train, y_train, color="blue")
plt.plot(x_train, x_pred, color="green")
plt.title("Salary vs Experience (Train)")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualization - Test
plt.scatter(x_test, y_test, color="blue")
plt.plot(x_train, x_pred, color="green")
plt.title("Salary vs Experience (Test)")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()

print("Accuracy: ", regressor.score(x_test, y_test) * 100)
```

```python
# Multiple Linear Regression - Advertising Sales
ad_df = pd.read_csv("advertising.csv")
x_ad = ad_df[['TV','Radio','Newspaper']]
y_ad = ad_df[['Sales']]

x_train, x_test, y_train, y_test = train_test_split(x_ad, y_ad, test_size=0.3, random_state=100)
mlr = LinearRegression()
mlr.fit(x_train, y_train)
print("Intercept:", mlr.intercept_)
print("Coefficients:", list(zip(x_ad.columns, mlr.coef_[0])))
```

```python
# Diabetes Dataset - sklearn
from sklearn.datasets import load_diabetes
X, Y = load_diabetes(return_X_y = True)
X = X[:, 0].reshape(-1,1)

lr = LinearRegression()
lr.fit(X,Y)
print("Intercept:", lr.intercept_)
print("Slope:", lr.coef_[0])

Y_pred = lr.predict(X)
plt.scatter(X, Y, label="Original Data")
plt.plot(X, Y_pred, color='r', label="Regression Line", linewidth=3)
plt.title("Diabetes Dataset - Regression")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.show()
```
