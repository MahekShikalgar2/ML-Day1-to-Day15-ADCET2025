```python
# Matplotlib Basics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simple Line Plot
xpoint = np.array([1,8])
ypoint = np.array([3,10])
plt.plot(xpoint, ypoint)
plt.title("Simple Line Plot")
plt.show()
```

```python
# Subplots with sin and cos
fig, ax = plt.subplots(2, 2, figsize=(10, 6))
x1 = np.linspace(0, 10, 100)
ax[0][0].plot(x1, np.sin(x1), 'g')
ax[0][1].plot(x1, np.cos(x1), 'y')
ax[1][0].plot(x1, np.sin(x1), 'b')
ax[1][1].plot(x1, np.cos(x1), 'r')
plt.tight_layout()
plt.suptitle("Sine and Cosine Subplots")
plt.show()
```

```python
# Markers and Lines
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o-.r')  # circle marker, dash-dot line, red
plt.title("Custom Marker and Line")
plt.show()
```

```python
# Marker size, edge, face color
plt.plot(ypoints, marker='o', ms=20, mec='m', mfc='y')
plt.title("Marker Styling")
plt.show()
```

```python
# Line Width
plt.plot(ypoints, linewidth=3.5)
plt.title("Line Width Example")
plt.show()
```

```python
# Multiple lines in one graph
y2 = np.array([6, 2, 7, 11])
plt.plot(ypoints)
plt.plot(y2)
plt.title("Multiple Lines")
plt.show()
```

```python
# Title and Labels
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse Rate")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()
```

```python
# Subplots Example
x = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([10,20,30,40])
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Sales")
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Income")
plt.suptitle("My Shopping Mall")
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot - Old and New Cars
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x1, y1, color='red')

x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x2, y2, color='#88c999')
plt.title("Old and New Car Speed Scatter Plot")
plt.show()
```

```python
# Scatter Plot with Colors
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.title("Colored Scatter Plot")
plt.show()
```

```python
# Scatter Plot with Size and Alpha
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title("Size-Based Scatter Plot")
plt.show()
```

```python
# Combined Color, Size, Alpha
x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
sizes = 10 * np.random.randint(100, size=100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.title("Random Colored & Sized Scatter")
plt.show()
```

```python
# Bar Chart
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
plt.title("Fruit Sales")
plt.show()

x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.barh(x, y)
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
# Custom Colors and Width
plt.bar(x, y, color="#4CAF50", width=0.5)
plt.title("Styled Bar Chart")
plt.show()
```

```python
# Histogram
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.title("Normal Distribution Histogram")
plt.show()
```

```python
# Styled Histogram
plt.style.use('_mpl-gallery')
x = 4 + np.random.normal(0, 1.5, 200)
fig, ax = plt.subplots(figsize=(8,6))
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")
ax.set(xlim=(0, 8), ylim=(0, 56))
plt.title("Styled Histogram")
plt.show()
```

```python
# Violin Plot
np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))
fig, ax = plt.subplots(figsize=(8,6))
vp = ax.violinplot(D, [2, 4, 6], widths=2, showmeans=False, showmedians=False, showextrema=False)
for body in vp['bodies']:
    body.set_alpha(0.9)
ax.set(xlim=(0, 8), ylim=(0, 8))
plt.title("Violin Plot")
plt.show()
```

```python
# Pie Charts
y = np.array([35, 25, 25, 15])
labels = ["Apples", "Bananas", "Cherries", "Dates"]
colors = ["red", "yellow", "hotpink", "brown"]
explode = [0.2, 0.2, 0.2, 0.2]
plt.pie(y, labels=labels, colors=colors, explode=explode, shadow=True, startangle=90)
plt.legend(title="Fruit Names")
plt.title("Fruit Pie Chart")
plt.show()
```

```python
# Hexbin Plot
plt.style.use('_mpl-gallery-nogrid')
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3
fig, ax = plt.subplots(figsize=(8, 8))
ax.hexbin(x, y, gridsize=20)
ax.set(xlim=(-2, 2), ylim=(-3, 3))
plt.title("Hexbin Plot")
plt.show()
```

```python
# Contour Plot
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)
fig, ax = plt.subplots(figsize=(6,6))
ax.contour(X, Y, Z, levels=levels)
plt.title("Contour Plot")
plt.show()
```

```python
# 3D Bar Plot
plt.style.use('_mpl-gallery')

x = [1, 1, 2, 2]
y = [1, 2, 1, 2]
z = [0, 0, 0, 0]
dx = np.ones_like(x)*0.5
dy = np.ones_like(x)*0.5
dz = [2, 3, 1, 4]

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(6,6))
ax.bar3d(x, y, z, dx, dy, dz)
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
plt.show()
```

```python
# 3D Scatter Plot
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))
ax.scatter(xs, ys, zs)
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
plt.show()
```

```python
# 3D Stem Plot
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

n = 20
x = np.sin(np.linspace(0, 2*np.pi, n))
y = np.cos(np.linspace(0, 2*np.pi, n))
z = np.linspace(0, 1, n)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))
ax.stem(x, y, z)
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
plt.show()
```

```python
# 3D Surface Plot
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
plt.style.use('_mpl-gallery')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(8,8))
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)
ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
plt.show()
```

```python
# Fill Between Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,100)
y1 = x**2
y2 = x

plt.fill_between(x,y1,y2, color='skyblue', alpha=0.4, label="Filled Area")
plt.plot(x,y1, label='y=x^2')
plt.plot(x,y2, label='y=x')
plt.title("Fill Between Example")
plt.legend()
plt.show()
```
