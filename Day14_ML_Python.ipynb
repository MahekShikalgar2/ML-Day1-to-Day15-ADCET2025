# Day-14: Dimensionality Reduction using PCA

```python
#Import required libraries
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
# Optional: Display the first few rows
print("Original Dataset (Iris):")
print(df.head())
```

```python
#  Standardize the data
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Optional: View the scaled data
print("\nScaled Data:")
print(scaled_data.head())
```

```python
# Heatmap before applying PCA
plt.figure(figsize=(6, 4))
sns.heatmap(scaled_data.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Scaled Features Before PCA")
plt.show()
```

```python
# Apply PCA
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
```

```python
#  Convert PCA result to DataFrame
data_pca = pd.DataFrame(pca_data, columns=['PC-1', 'PC-2', 'PC-3'])
```

```python
# Optional: View PCA-transformed data
print("\nPCA Transformed Data:")
print(data_pca.head())
```

```python
#  Heatmap of PCA components
plt.figure(figsize=(6, 4))
sns.heatmap(data_pca.corr(), annot=True, cmap='YlGnBu')
plt.title("Heatmap After PCA")
plt.show()
```
