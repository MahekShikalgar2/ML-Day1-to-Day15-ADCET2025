```python
#Input and Type Conversion
num = input("Enter a number: ")
print("You entered:", num)
print("Type:", type(num))
```

```python
# Explicit Conversion
num_int = int(num)
print("Converted to int:", type(num_int))
```

```python
# Arithmetic Operators
print("Addition:", 10 + 5)
print("Division:", 10 / 3)
print("Floor Division:", 10 // 3)
print("Exponentiation:", 2 ** 4)
```

```python
# Conditional Statements
num = -4
if num > 0:
    print("Positive number")
elif num == 0:
    print("Number is zero")
else:
    print("Negative number")
```

```python
# For Loop with List
fruits = ["apple", "mango", "orange"]
for fruit in fruits:
    print(fruit)
```

```python
# For Loop with String
for x in 'Python':
    print(x)
```

```python
# While Loop
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# Data Types
integer_num = 123
float_num = 1.23
new_num = integer_num + float_num
print("Sum:", new_num)
print("Datatype:", type(new_num))
```

```python
# Boolean and None
PI = 3.14
GRAVITY = 9.8
result1 = True
value = None
print("PI:", PI)
print("Gravity:", GRAVITY)
print("Boolean Result:", result1)
print("None Type:", value)
```

```python
# Data Structures
fruits = ["apple", "mango", "orange"]
print(fruits)
numbers = (1, 2, 3)
print(numbers)
alphabets = {'a': 'apple', 'b': 'ball', 'c': 'cat'}
print(alphabets)
vowels = {'a', 'e', 'i', 'o', 'u'}
print(vowels)
```

```python
# Logical Operators
a = 5
b = 6
print("Logical AND:", (a > 2) and (b >= 6))
print("Logical OR:", (a > 2) or (b >= 6))
print("Logical NOT:", not a)
```

```python
# Bitwise Operators
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1, 2, 3]
y3 = [1, 2, 3]
print("x1 is not y1:", x1 is not y1)
print("x2 is y2:", x2 is y2)
print("x3 is not y3:", x3 is not y3)
```

```python
# Total sum using while loop until user enters 0
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter another number (0 to stop): "))
print("Total Sum =", total)
```
