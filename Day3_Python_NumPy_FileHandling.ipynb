```python
# Mandelbrot Plot (Optional)
def mandelbrot(h, w, maxit=20):
    y, x = np.ogrid[-1.4:1.4:h*1j, -2:0.8:w*1j]
    c = x + y*1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)
    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2
        div_now = diverge & (divtime == maxit)
        divtime[div_now] = i
        z[diverge] = 2
    return divtime

plt.imshow(mandelbrot(400, 600))
plt.title("Mandelbrot Fractal")
plt.show()
```

```python
# Basic try-except
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print(result)
except:
    print("Error: Denominator cannot be zero")
```

```python
# Handling specific exceptions
try:
    even_numbers = [2, 4, 6, 8]
    print(even_numbers[5])
except ZeroDivisionError:
    print("Denominator cannot be zero")
except IndexError:
    print("Index is out of bound")
```

```python
# Assertion and else block
try:
    num = int(input("Enter a number: "))
    assert num % 2 == 0
except:
    print("Not an even number")
else:
    reciprocal = 1 / num
    print("Reciprocal:", reciprocal)
```

```python
# try-except-finally
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print(result)
except:
    print("Error: Denominator should be changed")
finally:
    print("This is finally block")
```

```python
# Custom Exceptions

class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass

try:
    input_num = int(input("Enter your age (in number): "))
    if input_num < 18:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
except InvalidAgeException:
    print("Exception/Error Occurred: InvalidAgeException")

class SalaryNotInRangeError(Exception):
    def __init__(self, salary, message="Salary is not in range (5000 to 15000)"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)

salary = int(input("Enter salary amount: "))
if not 5000 < salary < 15000:
    raise SalaryNotInRangeError(salary)
```
