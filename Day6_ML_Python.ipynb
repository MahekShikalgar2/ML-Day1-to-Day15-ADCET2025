```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Seaborn Version:", sns.__version__)
```

```python
# Load built-in dataset
df = sns.load_dataset('tips')
print(df.head())
```

```python
# Custom sinplot
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)

sns.set()
sinplot()
plt.title("Basic Sinplot using Seaborn")
plt.show()
```

```python
# Themes
for style in ["darkgrid", "whitegrid", "white", "ticks"]:
    sns.set_style(style)
    sinplot()
    plt.title(f"Seaborn style: {style}")
    sns.despine()
    plt.show()
```

```python
# Color Palettes
sns.palplot(sns.color_palette())  # Default
plt.show()

sns.palplot(sns.color_palette("Greens"))  # Sequential
plt.show()

sns.palplot(sns.color_palette("BrBG", 7))  # Diverging
plt.show()
```

```python
# Histogram (Seaborn)
sns.histplot(df['tip'], kde=False)
plt.title("Tip Distribution")
plt.show()
```

```python
#  Scatter Plot (Iris)
iris = sns.load_dataset('iris')
sns.jointplot(x='petal_length', y='petal_width', data=iris)
plt.show()
```

```python
# KDE Plot
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='kde')
plt.show()
```

```python
# Pair Plot
sns.pairplot(iris, hue='species', diag_kind="kde", palette="husl")
plt.show()
```

```python
# Strip & Swarm
sns.stripplot(x="species", y="petal_length", data=iris)
plt.title("Stripplot")
plt.show()
```

```python
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True)
plt.title("Stripplot with Jitter")
plt.show()

```

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.title("Swarmplot")
plt.show()
```

```python
# Box and Violin
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Boxplot")
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot - Tips")
plt.show()
```
